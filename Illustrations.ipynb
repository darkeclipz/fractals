# Number in the complex plane

```python
P = np.array([3, 2])
lim = 10
plt.figure(figsize=(5,5))
plt.plot(P[0], P[1], 'o', c='r')
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)
plt.axhline(0, ls='dashed', c='black', alpha=0.33)
plt.axvline(0, ls='dashed', c='black', alpha=0.33)
plt.xlabel('Real component')
plt.ylabel('Imaginary component')
plt.text(P[0] + lim/20, P[1] - lim/10, '{} + {}i'.format(*P))
```

# Diverging orbit

```python
c = complex(0.39, 0.4)
lim = 2
plt.figure(figsize=(5,5))
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)
plt.axhline(0, ls='dashed', c='black', alpha=0.33)
plt.axvline(0, ls='dashed', c='black', alpha=0.33)
plt.xlabel('Real component')
plt.ylabel('Imaginary component')
plt.text(c.real + lim/20, c.imag - lim/10, '{} + {}i'.format(c.real, c.imag))
orbit = []
xs = []; ys = [];
z = complex(0)
for i in range(12):
    z = z**2 + c
    xs.append(z.real)
    ys.append(z.imag)
plot(xs, ys, c='gray')
plot(xs, ys, 'o', c='gray', ms=3)
plt.plot(c.real, c.imag, 'o', c='r')
```
